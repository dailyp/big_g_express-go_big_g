##### Solution 2: Use Haversine method (with numpy)

```python
# https://www.geeksforgeeks.org/program-distance-two-points-earth/
# Python 3 program to calculate Distance Between Two Points on Earth
from math import radians, cos, sin, asin, sqrt

def get_distance_in_km(lats1, longs1, lats2, longs2):
    """Takes a np.array of coordinates and calculate the distance between 2 coordinates in Km using the Harvesine method"""
     
    # The numpy library contains a function named np.radians() 
    # which converts an array of floats from degrees to radians.
    lats1 = np.radians(lats1)
    lats2 = np.radians(lats2)
    longs1 = np.radians(longs1)
    longs2 = np.radians(longs2)
      
    # Apply Haversine formula
    delta_longs = longs2 - longs1
    delta_lats = lats2 - lats1
    a = np.sin(delta_lats / 2)**2 + np.cos(lats1) * np.cos(lats2) * np.sin(delta_longs / 2)**2
 
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Radius of earth in kilometers. Use 3956 for miles
    r = 6371
      
    # calculate the result
    return(c * r)
```
